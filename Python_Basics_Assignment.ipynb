2.	Explain the difference between List and Tuple in Python.

Ans.
•	List is mutable (can be changed), while Tuple is immutable (cannot be changed).

•	Lists use square brackets [ ], while Tuples use round brackets ( ).

•	Lists are slower, while Tuples are faster and more secure.
